```python
import matplotlib.pyplot as plt
import numpy as np

def plot_toilet_time():
    x = np.linspace(0, 10, 100)  # x values from 0 to 10 minutes
    n_values = [1, 2, 3, 4, 5]  # n values (number of toilets)
    
    plt.figure(figsize=(8, 6))
    for n in n_values:
        y = x * (n + 1) / 2  # Updated average toilet time formula
        # Add the annotation with the calculation
        plt.plot(x, y, label=None)  
        plt.annotate(f'{n} toilet{"s" if n > 1 else ""}', 
                     xy=(8, 8 * (n + 1) / 2), xytext=(8, 8 * (n + 1) / 2),
                     fontsize=10, color='black', ha='center', va='bottom')
    
    #plt.title("Time spend on the throne")
    plt.xlabel("Wait time (when first in line)")
    plt.ylabel("Average Time")
    plt.xlim(0, 10)  # Set x-axis limits to include zero
    plt.ylim(0, 30)  # Adjusted to match the new formula's range
    plt.grid(True)
    plt.savefig("static/toilet_time.png")
    plt.show(True)
    plt.close()

plot_toilet_time()
```
